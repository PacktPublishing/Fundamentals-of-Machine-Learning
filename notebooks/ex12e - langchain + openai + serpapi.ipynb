```python
from langchain.agents import load_tools # pip install -U langchain-community
from langchain.agents import initialize_agent
from langchain.agents import AgentType
from langchain.llms import OpenAI
```

```python
from google.colab import userdata

OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')
SERPAPI_API_KEY = userdata.get('SERPAPI_API_KEY')
```

```python
import os
os.environ["OPENAI_API_KEY"] = OPENAI_API_KEY
os.environ["SERPAPI_API_KEY"] = SERPAPI_API_KEY
```

```python
llm = OpenAI(temperature=0)
tools = load_tools(["serpapi","llm-math"], llm=llm)
```

```python
agent = initialize_agent(tools, llm, agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION, verbose=True)
```

```python
output = agent.run("who is that actor in 2023 oscar? what is his age raised to the 0.5 power?")
```

```python
output = agent.run("who is Yiqiao Yin? How long has he been in his career? What's the number of years of experience of his career? Take that number and multiply it by 10.")
```
